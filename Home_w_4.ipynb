#### 1.3 Изучите распределение квадратуры жилой


```python
plt.figure(figsize=(10, 6))

plt.hist(df['sqft_living'], bins=80)

plt.title('Распределение жилой квадратуры')
plt.xlabel('Кв. метры жилые')
plt.ylabel('Количество домов');
```

#### 1.4 Изучите распределение года постройки

## Задание 2

#### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график. Сделайте выводы.


```python
water_view=df['waterfront'].value_counts()
water_view.head()
```

```python
plt.figure(figsize=(10, 4))

plt.pie(water_view.values, autopct='%1.1f%%')

plt.title('Вид на набережную')
plt.legend(['нет', 'есть']);
```

Домов с видом на набережную меньше 1%

#### 2.2 Изучите распределение этажей домов


```python
floors_c=df['floors'].value_counts()
floors_c.head()
```

```python
plt.figure(figsize=(8, 8))

plt.pie(floors_c.values, autopct='%1.1f%%')

plt.legend(floors_c.index);
```

Больше всего 1-этажных домов

#### 2.3 Изучите распределение состояния домов


```python
condition_1=df['condition'].value_counts()
condition_1
```

```python
plt.figure(figsize=(10, 5))

plt.bar(condition_1.index, condition_1.values)

plt.title('Распределение состояния домов')
plt.xlabel('Состояние домов')
plt.ylabel('Количество домов');
```

Большее количество домов находится в удовлетворительном состоянии
